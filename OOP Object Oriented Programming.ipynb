```python
class Restaurant():
    def __init__(self, restaurant_name, cuisine_type):
        self.restaurant_name = restaurant_name
        self.cuisine_type = cuisine_type

    def describe_restaurant(self):
        print(f"Restaurant name: {self.restaurant_name}, Cuisine Type: {self.cuisine_type}")

    def open_restaurant(self):
        print("Restaurant is open")

restaurant = Restaurant("Seth jee", "Desi")
restaurant.describe_restaurant()
restaurant.open_restaurant()
print(restaurant.restaurant_name)
```

```python
class Student():
    def __init__(self,name,marks):
        self.name = name
        self.marks = marks

    def avg(self):
        sum = 0
        for i in self.marks:
            sum += i

        avg = sum/3
        print(f"The average marks of {self.name} is {avg}")

student1 = Student("Amin",[90,80,70])
student1.avg()

student2 = Student("Ayesha",[50,50,50])
student2.avg()
```

# inheitance

```python
# Inheritance is a mechanism where one class (called the child class) inherits attributes and methods from another class (called the parent class). This promotes code reusability and logical hierarchy.
class Car:
  color = "black"
  @staticmethod
  def start():
    print("Car started")

  @staticmethod
  def stop():
    print("Car Stop")

class ToyotaCar(Car):
  def __init__(self,name):
    self.name = name

car1 = ToyotaCar("fortuner")
car2 = ToyotaCar("prius")


print(ToyotaCar.start())
print(ToyotaCar.stop())

print(car1.name)
print(car2.name)

print(car1.start())
print(car2.stop())

car1.color
```

```python
class Car:
  color = "black"
  @staticmethod
  def start():
    print("Car started")

  @staticmethod
  def stop():
    print("Car Stop")

class ToyotaCar(Car):
  def __init__(self,brand):
    self.brand = brand

class Fortuner(ToyotaCar):
  def __init__(self,type):
    self.type = type

car1 = Fortuner("petrol")
car2 = Fortuner("deiasal")


print(car1.start())
print(car2.stop())

print(car2.type)
```

### super method

```python
# super method():
# super() function is used to give access to methods and properties of a parent or sibling class.

class Car:
  def __init__(self,type):
    self.type = type
  @staticmethod
  def start():
    print("Car started")
  @staticmethod
  def stop():
    print("Car Stop")

class ToyotaCar(Car):
  def __init__(self,name,type):
    self.name = name
    super().__init__(type)
    super().start()
    super().stop()

car1 = ToyotaCar("fortuner","petrol")

print(car1.type)
print(car1.type)
```

```python
class Car:
  def __init__(self,type):
    self.type = type
  @staticmethod
  def start():
    print("Car started")
  @staticmethod
  def stop():
    print("Car Stop")

class ToyotaCar(Car):
  def __init__(self,name,type):
    super().__init__(type)
    self.name = name
    super().start()
    super().stop()

car1 = ToyotaCar("fortuner","petrol")

print(car1.type)
print(car1.type)
```
